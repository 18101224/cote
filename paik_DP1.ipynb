<h1>피보나치수열</h1>

```python
dp = [1,1]
def fib(n):#dp
    if n <= len(dp):
        return dp[n-1]
    elif n == len(dp)+1 : 
        dp.append(dp[n-3]+dp[n-2])
        return dp[-1]
    else:
        k=fib(n-1)+fib(n-2)
        dp.append(k)
        return k
```

```python
fib(21)
```

```python
def fib2(n):#재귀호출
    if n ==1 or n==2:
        return 1
    else:
        return fib2(n-1)+fib2(n-2)
```

```python
fib2(21)
```

```python
dp = [1,2,3,5]
def f(n):
    if n == 1:
        return 1
    if n == 2:
        return 2
    if n==3 :
        return 3
    if n==4:
        return 5
    if len(dp) < n-2 :
        dp.append(f(n-2))
    if len(dp) < n-1 :
        dp.append(f(n-1))
    return dp[n-3]+dp[n-2]
```
